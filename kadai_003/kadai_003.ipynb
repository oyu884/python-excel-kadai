```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill

# Step1 Excel読み込み
df_2022 = pd.read_excel("2022_年間売上表.xlsx")
df_2023 = pd.read_excel("2023_年間売上表.xlsx")

# Step2 データ結合
df = pd.concat([df_2022, df_2023], ignore_index=True)

# Step3 商品・売上年ごとに集計
summary_df = df.groupby(["商品", "売上年"], as_index=False)["金額（千円）"].sum()

# Excel書き込み
file_name = "売上集計表.xlsx"
summary_df.to_excel(file_name, index=False)

# Step4 openpyxlで見た目調整
wb = load_workbook(file_name)
ws = wb.active

fill = PatternFill(start_color="F2F2F2", end_color="F2F2F2", fill_type="solid")
font = Font(bold=True)

for cell in ws[1]:
    cell.fill = fill
    cell.font = font

wb.save(file_name)
```
